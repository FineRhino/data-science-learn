## Manipulating Data 2

```python
# Create a column from a series
seats_column = pd.Series([5,5,5,5,5])

# New column called seats - by default, added at end
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
```

```python
car_sales
```

```python
# Column from a Python list - NOTE - The number of items in a List (unlike a Series, must match the number of rows)
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 8.7, 3.0, 4.5, 9.0]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
# Determine how many liters of fuel a car has used in its entire life
car_sales["Total Fuel Used (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
```

```python
car_sales
```

```python
# Create a coumn from a single value
car_sales["Number of Wheels"] = 4
```

```python
car_sales
```

```python
car_sales["Passed Safety"] = True
car_sales
```

```python
car_sales.dtypes
```

```python
car_sales = car_sales.drop("Total Fuel Used", axis=1)
car_sales
```

## Manipulating Data - 3

```python
# Shuffling Data
car_sales_shuffle = car_sales.sample(frac=1) # 100% of the data will be shuffled
car_sales_shuffle
```

```python
# Only select 20% of the data
car_sales_shuffle_20 = car_sales.sample(frac=.2)
car_sales_shuffle_20
```

```python
# Reset Index
car_sales_shuffle.reset_index(inplace=True, drop=True)
print(car_sales_shuffle)
```

```python
# Assign a function to a column using the apply function
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x / 1.6) # lambda is an anonymous function; km:miles is 1:1.6
car_sales
```
