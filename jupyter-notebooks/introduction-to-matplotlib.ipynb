# Matplotlib Example Wrokflow

```python
# 0. Import matplotlib and get it ready for plotting in Juptyer
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10, 10))

# 3. Plot the Data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save and show (you save the whole figure)
fig.savefig("sample-plot.png")
```

```python
# Make a bar plot from a dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Dan's Nut Butter Store",
       xlabel="Nut Butter Type",
       ylabel="Price ($)");
```

```python
# Horizontal Bar
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Plotting from Pandas DataFrames - 5

### Which one should you use? (pyplot vs matplotlib OO method?)
* When plotting something quickly, ok to use the pyplot method
* When plotting somehting more advanced, use the OO method

```python
heart_disease_over_50 = heart_disease[heart_disease["age"] > 50]
heart_disease_over_50.head()
```

```python
# pyplot method
heart_disease_over_50.plot(kind="scatter",
                           x='age',
                           y='chol',
                           c='target');
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10, 6))
heart_disease_over_50.plot(kind='scatter',
                           x='age',
                           y='chol',
                           c='target',
                           ax=ax);
#ax.set_xlim([45, 100]);
```

## Plotting from Pandas DataFrames - 6

```python
heart_disease_over_50.target.values
```

```python
## OO Method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=heart_disease_over_50["age"],
                     y=heart_disease_over_50["chol"],
                     c=heart_disease_over_50["target"]);

# Customize
ax.set(title="Heart Disease and Cholestoerol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a legend
# NOTE - The below is not working, although it should. See https://matplotlib.org/stable/gallery/lines_bars_and_markers/scatter_with_legend.html
# ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line - axhline
ax.axhline(heart_disease_over_50["chol"].mean(),
           linestyle='--');
```

## Plotting from Pandas DataFrames 7

```python
import matplotlib
heart_disease_over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1,
                               figsize=(10, 10));

# Add data to ax0
scatter = ax0.scatter(x=heart_disease_over_50["age"],
                      y=heart_disease_over_50["chol"],
                      c=heart_disease_over_50["target"]);

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol");

# Add a legend to ax0
ax0.legend(title="Target")

# Add a mean line to ax0
ax0.axhline(heart_disease_over_50["chol"].mean(),
            linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=heart_disease_over_50["age"],
                      y=heart_disease_over_50["thalach"],
                      c=heart_disease_over_50["target"]);

# Customize ax1
ax1.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Max Heart Rate");

# Add a legend to ax1
ax1.legend(title="Target")

# Add a meanline to ax1
ax1.axhline(heart_disease_over_50["thalach"].mean(),
            linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

## Customizing Your Plots - 2

```python
plt.style.use('seaborn-whitegrid')

## OO Method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=heart_disease_over_50["age"],
                     y=heart_disease_over_50["chol"],
                     c=heart_disease_over_50["target"],
                     cmap="winter"); #this changes the color scheme

# Customize
ax.set(title="Heart Disease and Cholestoerol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a legend
# NOTE - The below is not working, although it should. See https://matplotlib.org/stable/gallery/lines_bars_and_markers/scatter_with_legend.html
# ax.legend(*scatter.legend_elements(), title="Target");
ax.legend(title="Target");

# Add a horizontal line - axhline
ax.axhline(heart_disease_over_50["chol"].mean(),
           linestyle='--');
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1,
                               figsize=(10, 10));

# Add data to ax0
scatter = ax0.scatter(x=heart_disease_over_50["age"],
                      y=heart_disease_over_50["chol"],
                      c=heart_disease_over_50["target"],
                      cmap="winter");

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol");

# Change the x axis limits
ax0.set_xlim([50, 80])

# Add a legend to ax0
ax0.legend(title="Target")

# Add a mean line to ax0
ax0.axhline(heart_disease_over_50["chol"].mean(),
            linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=heart_disease_over_50["age"],
                      y=heart_disease_over_50["thalach"],
                      c=heart_disease_over_50["target"],
                      cmap="winter");

# Customize ax1
ax1.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Max Heart Rate");

# Change the x axs limits
ax1.set_xlim([50, 80])
# Change the y axis limits
ax1.set_ylim([60, 200])

# Add a legend to ax1
ax1.legend(title="Target")

# Add a meanline to ax1
ax1.axhline(heart_disease_over_50["thalach"].mean(),
            linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
